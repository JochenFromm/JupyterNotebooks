```python
from numpy import log2
import numpy as np
import matplotlib.pyplot as plt

b = 5
N = 20
k = int(log2(b)*N)+1

for i in range(N):
    s = format(b**i, str(k).zfill(N)+'b')
    print(s)
```

```python
import matplotlib.cm as cm

N = 200
z = 350
w = 100
h = 100

def binary_raster(b):
  r = []
  for i in range(N):
    s = bin(b**i)[2:].zfill(z)    
    bits = [1-int(c) for c in s[-z:]]
    r.append(bits[-z:])
  return r
```

```python
dpi = 12
base = 3
plt.figure(figsize=(w, h), dpi=dpi)
plt.imshow(binary_raster(base), cmap=cm.gray)
plt.savefig(f"exponentiation_base_{base}.png")
plt.show()
```

```python
dpi = 12
base = 5
plt.figure(figsize=(w, h), dpi=dpi)
plt.imshow(binary_raster(base), cmap=cm.gray)
plt.savefig(f"exponentiation_base_{b}.png")
plt.show()
```

```python
N = 200
z = 350
w = 100
h = 100

def hex_raster(b):
  r = []
  for i in range(N):
    s = hex(b**i)[2:].zfill(z)    
    bits = [16-int(c, 16) for c in s[-z:]]
    r.append(bits[-z:])
  return r
```

```python
dpi = 12
base = 3
plt.figure(figsize=(w, h), dpi=dpi)
plt.imshow(hex_raster(base), cmap=cm.viridis)
plt.savefig(f"exponentiation_base_{base}_16.png")
plt.show()
```
